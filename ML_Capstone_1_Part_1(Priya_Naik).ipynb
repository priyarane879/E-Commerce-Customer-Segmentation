### 2. Data preprocessing

2.1 Checking for Duplicate Value

```python
# Dataset Duplicate Value Count
#Checking for duplicate rows in dataframe
len(df_customer[df_customer.duplicated()])
```

```python
df_customer.isna().sum() #checking for null value
```

* As there are 4223 type of unique description and if we will replace null value with some description then this will add noise in our original dataset so it's better to remove that rows.

2.2.2 Handing Missing Value in CustomerID Column.

```python
# This will gives us no. of occurance of each unique value from column children.
df_customer['CustomerID'].value_counts()
```

Observation:-
* As customerID is unique to each customer , so If I will replace null values with one particular ID then it will create noise in my dataset, so I will remove the rows with missing customer ID.

```python
df_customer.dropna(inplace=True)# dropping those rows which contain null values.
```

2.3 Checking for Outlier in Numerical Variable

```python
df_customer['Quantity'].value_counts()
```

Observation:-
* As quantity can not be zero , I will select those rows where quantity is greater than zero.

```python
df_customer=df_customer[(df_customer['Quantity']>0)] # keeping those rows whose quantity column value greater than 0
```

2.4 Changing inappropriate datatype

```python
df_customer['InvoiceNo'] = df_customer['InvoiceNo'].apply(int)
df_customer['InvoiceDate'] = df_customer['InvoiceDate'].apply(pd.to_datetime)
df_customer['CustomerID'] = df_customer['CustomerID'].apply(int)
```

```python
df_customer.info()
```

### 3.Feature Engineering & EDA

####3.1 Feature Engineering

3.1.1 Adding New Column

```python
# Creating a new column named 'Amount' in the  dataframe
df_customer['Amount'] = df_customer['Quantity']*df_customer['UnitPrice']
# Creating a new column named 'MMonth' in the  dataframe
df_customer['Month'] = df_customer['InvoiceDate'].dt.strftime("%B")
# Creating a new column named 'Year' in the  dataframe
df_customer['Year'] = df_customer['InvoiceDate'].dt.year
# Creating a new column named Time in the data frame
df_customer['Time']=df_customer['InvoiceDate'].dt.time
# Creating a new column named Hour in the data frame
df_customer['Hour']=df_customer['InvoiceDate'].dt.hour
# Creating a new column named 'Day' in the dataframe
df_customer['Day'] = df_customer['InvoiceDate'].dt.day_name()

#changing datatype to integer
df_customer['Year'] = df_customer['Year'].apply(int)
```

Observation:-
* Added new columns Month, Year,Time, Hour, Day,& 'Amount' which is product of quantity and price.
